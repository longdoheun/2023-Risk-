```python
# Step 2. Complement code that implements backtesting for value weight portfolio

date_index = df_ret.index.copy().tolist()
for cnt, idx in enumerate(df_ret.index):
    if cnt == 0:
        continue

    else:
        target_weight = df_mktcap.loc[date_index[cnt-1]] / df_mktcap.loc[date_index[cnt-1]].sum()
        current_weight = (1+df_ret.loc[date_index[cnt-1]]) * df_weight.loc[date_index[cnt-1]]
        current_weight = current_weight / current_weight.sum()
        weight_difference = np.abs(target_weight - current_weight)

        df_weight.loc[date_index[cnt]] = target_weight
        pf_ret.loc[date_index[cnt],"pf_value"]  = (df_ret.loc[date_index[cnt]] * df_weight.loc[date_index[cnt]]).sum() - (weight_difference.sum() * 0.002)
```

```python
# Step 3. Scalable Code
# def generate_weight_func():
#     ~ ~ ~
#     return weight


def generate_mktcap_weight(date_index, when, mktcap_df):
    """
    date_index (lits): date index
    when (int): 0 for now, - for past, + for future
    mktcap_df (DataFrame): market capitalization dataframe
    """

    weight = mktcap_df.loc[date_index[when]] / mktcap_df.loc[date_index[when]].sum()

    return weight

date_index = df_ret.index.copy().tolist()
for cnt,idx in enumerate(df_ret.index):
    if cnt == 0:
        continue

    else:
        target_weight = generate_mktcap_weight(date_index, cnt-1, df_mktcap)
        current_weight = (1+df_ret.loc[date_index[cnt-1]]) * df_weight.loc[date_index[cnt-1]]
        current_weight = current_weight / current_weight.sum()
        weight_difference = np.abs(target_weight - current_weight)

        df_weight.loc[date_index[cnt]] = target_weight
        pf_ret.loc[date_index[cnt],"pf_value"]  = (df_ret.loc[date_index[cnt]] * df_weight.loc[date_index[cnt]]).sum() - (weight_difference.sum() * 0.002)
```

```python
# Step 3. Scalable Code -- With minimum varaince
# def generate_weight_func():
#     ~ ~ ~
#     return weight

def generate_minvar_weight(date_index, when, df_ret):
    returns = df_ret.loc[:date_index[when]].mean(axis=0).values
    covariance = df_ret.cov().values

    I = np.ones_like(returns)
    weight = np.linalg.solve(covariance, I)
    weight = weight / np.sum(weight)

    return weight

date_index = df_ret.index.copy().tolist()
for cnt,idx in enumerate(df_ret.index):
    if cnt == 0:
        continue

    elif idx < pd.to_datetime("2020"):
        continue

    else:
        target_weight = generate_minvar_weight(date_index, cnt-1, df_ret)
        current_weight = (1+df_ret.loc[date_index[cnt-1]]) * df_weight.loc[date_index[cnt-1]]
        current_weight = current_weight / current_weight.sum()
        weight_difference = np.abs(target_weight - current_weight)

        df_weight.loc[date_index[cnt]] = target_weight
        pf_ret.loc[date_index[cnt],"pf_value"]  = (df_ret.loc[date_index[cnt]] * df_weight.loc[date_index[cnt]]).sum() - (weight_difference.sum() * 0.002)
```
