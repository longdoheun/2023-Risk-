```python
### Backtesting with for loop - Value ###
# Implement Value weight portfolio using for loop
```

```python
# Step 1. Before starting the loop, define weight dataframe and portfolio return dataframe
```

```python
df_weight = pd.DataFrame(np.zeros_like(df_ret), index=df_ret.index, columns=df_ret.columns)
pf_ret = pd.DataFrame(np.zeros(len(df_ret)), index=df_ret.index, columns=["pf_ret"])
```

```python
# Step 2. Complement code that implements backtesting for value weight portfolio
date_index = df_ret.index.copy().tolist()
for cnt,idx in enumerate(df_ret.index):
    if cnt == 0:
        continue

    current_weight = (1+df_ret.loc[date_index[cnt-1]]) * (df_weight.loc[date_index[cnt-1]])
    target_weight = df_mktcap.loc[date_index[cnt-1]] / df_mktcap.loc[date_index[cnt-1]].sum()

    weight_difference = np.abs(target_weight - current_weight)

    df_weight.loc[idx] = target_weight
    pf_ret.loc[idx] = (df_weight.loc[idx] * df_ret.loc[idx]).sum()
```
